Placement Prediction using Linear Regression

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse, mean_absolute_error as mae, r2_score
```

# Loading the dataset

```python
df = pd.read_csv('placement.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x = 'cgpa', y = 'package', data = df);
```

# Correlation

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True);
```

# Seggrigate the data x and y

```python
x = df[['cgpa']]
y = df['package']
```

# Splitting the data into training data and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.20, random_state = 1)
```

```python
x_train
```

```python
y_train
```

```python
x_test.shape
```

```python
x_test
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
c = lr.intercept_ # y intercept
c
```

```python
m = lr.coef_ # slope (m)
m
```

# Evaluating the model performance

```python
y_pred_train = lr.predict(x_train)
y_pred_train
```

```python
y_pred_test = lr.predict(x_test)
y_pred_test
```

```python
r2 = r2_score(y_test, y_pred_test)
r2
```

from this r2 score we can say that your cgpa which is your input column is able to explain the 72% of varience in the package column

```python
MAE = mae(y_test, y_pred_test)
MAE
```

```python
MSE = mse(y_test, y_pred_test)
MSE
```

```python
rmse = np.sqrt(MSE)
rmse
```

```python
plt.scatter(df['cgpa'], df['package']);
plt.plot(x_train['cgpa'], lr.predict(x_train), color='red')
```
